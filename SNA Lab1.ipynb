## Solution:


```python
GF = nx.complete_graph(5)
nx.draw_networkx(GF)
plt.show()
```

- The density for undirected graphs is:
$$d = \frac{2m}{n(n-1)}$$
- The density for directed graphs is:
$$d = \frac{m}{n(n-1)}$$
where \(n\) is the number of nodes and \(m\) is the number of edges in \(G\).

- The density of nodes in general is :
$$d_{n} = \frac{n(n-1)}{P_{c}}$$
where (n) is the number of nodes and (P_{c}) is the number of the number of potential connections.

```python
nx.density(GF)
```

```python
def Centrality_measures(GF):
    print('Degree centrality:', nx.degree_centrality(GF))
    #print('In degree centrality:', nx.in_degree_centrality(GF)) # For directed graph
    #print('Out degree centrality:', nx.out_degree_centrality(GF)) # For directed graph
    print('Beteweenness centrality:', nx.betweenness_centrality(GF))
    #print('Current-flow closeness centrality', nx.current_flow_closeness_centrality(GF, solver='lu'))
    #print('Information centrality', nx.information_centrality(GF, solver='lu'))
Centrality_measures(GF)
```

```python
#np.random.seed(0)
Edges = list(GF.edges())
print(Edges)
while GF.number_of_edges() > 1:
    i = random.randint(len(list(GF.edges())))
    print('Edge to be removed from the graph is', list(GF.edges())[i])
    #GF.remove_node(i)
    GF.remove_edge(list(GF.edges())[i][0], list(GF.edges())[i][1])
    
    Centrality_measures(GF)
    nx.draw_networkx(GF)
    plt.show()
```
